# 2. Define functions
- Key elements involving in the gradient descent are: 
    - Cost function: Such as MSE for linear regression.
    $$ J(\beta) = \frac {1} {2m} \sum_{i=1}^{m} (x\beta - y)^2 $$
    - Derivatives respect to the coefficients of predictors: Slope of the gradient.
    $$ \frac {\partial J(\beta)} {\partial \beta} = \frac {1} {m} \sum_{i=1}^{m} (x\beta - y) x $$
    - Learning rate: How large a step needed to update the gradient.
    - How to update beta(s)
    $$ \beta^{t+1} = \beta^{t} - \eta \frac {\partial J(\beta)} {\partial \beta} $$
- Steps (by each iteration)
    - 1. Given a set of betas (coefficients), compute the prediction.
    - 2. Update a set of betas (by the gradient).

```python
def cost_function(beta, x, y):
    '''
    beta: vector of betas, the shape must be (number of predictors, 1) 
    x: matrix of x with added bias units (beta 0), the shape must be 
    (number of data, number of predictors)
    y: vector of y, the shape must be (number of data, 1)
    
    Given beta(s), the coefficients of predictors, x (predictors), and y 
    (the ground truth values), return the corresponding cost 
    (the distance between predictions and the ground truth values).
    '''
    # number of data points
    m = len(y)
    
    # prediction
    prediction = np.dot(x, beta)
    
    # cost function for gradient descent
    cost = (1/m) * sum((prediction - y)**2)
    
    return cost
```

```python
def gradient_descent(x, y, beta, learning_rate=0.1, iterations=1000):
    '''
    x: matrix of x with added bias units (beta 0), the shape must be 
    (number of data, number of predictors)
    y: vector of y, the shape must be (number of data, 1)
    beta: vector of betas, the shape must be (number of predictors, 1) 
    learning_rate: a float number 
    iterations: number of iterations
    
    Given beta(s), the coefficients of predictors, x (predictors), y (ground truth),
    learning rate, and the number of iterations, return 'updated' beta(s) which 
    achieve the minimum cost on the data.
    
    Also, return the cost and beta history for latter visualization purposes.
    '''
    
    # number of data points
    m = len(y)
    
    # record the cost history 
    cost_history = np.zeros(iterations)
    
    # testing two betas (beta 0 & beta 1)
    beta_history = np.zeros((iterations, 2))
    
    # start updating the gradient for each beta 
    for i in range(iterations):
        prediction = np.dot(x, beta)
        beta = beta - (1/m)*learning_rate*(np.dot(x.T, prediction-y))
        beta_history[i, :] = beta.T
        cost_history[i] = cost_function(beta=beta, x=x, y=y)
        
    return beta, cost_history, beta_history
```

# 3. Generate data

```python
# Generate some random data for testing programs
X = 2*np.random.rand(100,1)
X_b = np.c_[np.ones((len(X), 1)), X]

y = 4+3*X+np.random.randn(100,1)
```

# 4. Traditonal way (using statsmodels)

```python
# Use statsmodels to fit the model again
import statsmodels.api as sm

# Fit the model on train data
model_1 = sm.OLS(y, X_b).fit()
model_1.summary()
```

```python
# Coefficients and MSE
print('Beta0:          {:0.3f},\nBeta1:          {:0.3f}'.format(model_1.params[0],
                                                                 model_1.params[1]))
print('Final cost/MSE:  {:0.3f}'.format(sum(model_1.resid**2)/100))
```

# 5. Gradient search 
- Use gradient search to find the estimated coefficients (beta 0 and beta 1), and compare the corresponding MSE with results attained from the traditional way.

```python
# Generate random starting betas 
start_beta = np.random.randn(2,1)

# Start gradient descent!
beta, cost_history, beta_history = gradient_descent(X_b, y,
                                                    beta=start_beta)
```

```python
# The results are all the same as using the traditional way!
print('Beta0:          {:0.3f},\nBeta1:          {:0.3f}'.format(beta[0][0],beta[1][0]))
print('Final cost/MSE:  {:0.3f}'.format(cost_history[-1]))
```

## 6.1 Searching history visualization 

```python
# Find which iteration has achieved the MSE close to the minimum MSE.
minimum_index = []
for i in range(len(cost_history)):
    if round(cost_history[i],3) ==  round(min(cost_history),3):
        minimum_index.append(i)
print(minimum_index[0])
```

```python
# It's clear that we are minimizing the MSE (cost function of linear regression)!
# The green dashed line indicates where I achieved the MSE close to the minimum MSE.
# Thus, we might not need too many iterations! 
fig,ax = plt.subplots(figsize=(10,7))
ax.set_title('MSE history (1000 iterations)', fontsize=20)
ax.set_ylabel('MSE', fontsize=15)
ax.set_xlabel('Iterations', fontsize=15)
_=ax.plot(range(1000),cost_history,'b.')
ax.vlines(minimum_index[0], ymin = min(cost_history), ymax = max(cost_history),
           linestyles = '--', linewidth = 4, colors = 'green')
```

```python
# We might only need around 330 iterations to reach the minimum MSE!
fig,ax = plt.subplots(figsize=(10,7))
ax.set_title("lr:{}, Iterations:{}, Min MSE:{}".format(0.1, 
                                                       1000,
                                                       round(min(cost_history),3)))
ax.set_ylabel('MSE', fontsize=15)
ax.set_xlabel('Iterations', fontsize=15)
_=ax.plot(range(minimum_index[0]+50),cost_history[:minimum_index[0]+50],'b.')
ax.vlines(minimum_index[0], ymin = min(cost_history), ymax = max(cost_history),
           linestyles = '--', linewidth = 4, colors = 'green')
```

## 6.2 Impact of different learning rates and iterations on the cost function (MSE)
- I have tested for combinations of learning rates and iterations to see how they affect the MSE.
    - For smaller learning rate, we might need to use more iterations to 'hit' the minimum MSE, which is time-consuming. 
    - For larger learning rate, we might 'pass' through the minimum MSE or even never hit the minimum MSE because we are taking a 'step' that is too big.
    - Therefore, there is no general rule to decide the learning rate and iterations.

```python
fig = plt.figure(figsize=(30,25),dpi=300)
fig.subplots_adjust(hspace=0.6, wspace=0.4)

it_lr =[(2000,0.001),(500,0.01),(200,0.05),(100,0.01)]
count = 0

for n_iter, lr in it_lr:
    count += 1
    ax = fig.add_subplot(4, 1, count)
    beta, cost_history, beta_history = gradient_descent(X_b, y, beta=start_beta,
                                                        learning_rate=lr,
                                                        iterations=n_iter
                                                        )
    minimum_index = []
    for i in range(len(cost_history)):
        if round(cost_history[i],3) ==  round(min(cost_history),3):
            minimum_index.append(i)
    
    ax.set_title("lr:{}, Iterations:{}, Min MSE:{}".format(lr, 
                                                           n_iter,
                                                           round(min(cost_history),3)), 
                 fontsize=20)
    ax.vlines(minimum_index[0], ymin = min(cost_history), ymax = max(cost_history),
           linestyles = '--', linewidth = 4, colors = 'green')
    
    _ = ax.plot(range(n_iter),cost_history,'b.')
```

# 7. Different types of gradient descent

## 7.1 Stochastic Gradient Descent
- Compared with the gradient descent which iterates through all the data, stochastic gradient descent iterates into a 'random' sample. 
- Steps (by each iteration)
    - 1. Select a random sample. 
    - 2. Compute the prediction given the selected random sample.
    - 3. Update a subset of betas. 
    - 4. Each iteration runs 'the number of sample size' times.

```python
def stocashtic_gradient_descent(x,y,beta,learning_rate=0.01,iterations=1000):
    '''
    x: matrix of x with added bias units (beta 0), the shape must be 
    (number of data, number of predictors)
    y: vector of y, the shape must be (number of data, 1)
    beta: vector of betas, the shape must be (number of predictors, 1) 
    learning_rate: a float number 
    iterations: number of iterations
    
    Returns the final beta vector and array of cost history.
    '''
    # number of data points
    m = len(y)
    
    # record the cost history (for latter visualization purpose)
    cost_history = np.zeros(iterations)
    
    # start iterations!
    for each_iter in range(iterations):
        
        for i in range(m):
            # for each iteration, extract a random sample
            rand_ind = np.random.randint(0,m)
            
            # the shape of x_i is always 1 by number of features
            x_i = x[rand_ind,:].reshape(1,x.shape[1])
            # the shape of x_i is always 1 by 1
            y_i = y[rand_ind].reshape(1,1)
            # prediction = x_i*beta(s)
            prediction = np.dot(x_i,beta)
            # update beta(s)
            beta = beta -(1/m)*learning_rate*(x_i.T.dot((prediction - y_i)))
            
        # record the cost once complete one iteration
        cost_history[each_iter]  = cost_function(beta,x,y)
        
        # to see the converge
        #if abs(cost_history[each_iter] - cost_history[each_iter-1]) <= 0.0001:
            #print(each_iter)
        
    return beta, cost_history
```

```python
# Start stochastic gradient descent!
beta, cost_history = stocashtic_gradient_descent(X_b, y,
                                                 beta=start_beta,
                                                 learning_rate=0.1,
                                                 iterations=1000)
```

```python
# Coefficients and MSE
print('Beta0:          {:0.3f},\nBeta1:          {:0.3f}'.format(beta[0][0],
                                                                 beta[1][0]))
print('Final cost/MSE:  {:0.3f}'.format(cost_history[-1]))
```

```python
# Find which iteration has achieved the MSE close to the minimum MSE.
minimum_index = []
for i in range(len(cost_history)):
    if round(cost_history[i],3) ==  round(min(cost_history),3):
        minimum_index.append(i)
print(minimum_index[0])
```

```python
# We might only need around 330 iterations to reach the minimum MSE!
fig,ax = plt.subplots(figsize=(10,7))
ax.set_title("lr:{}, Iterations:{}, Min MSE:{}".format(0.1, 
                                                       1000,
                                                       round(min(cost_history),3)))
ax.set_ylabel('MSE', fontsize=15)
ax.set_xlabel('Iterations', fontsize=15)
_=ax.plot(range(minimum_index[0]+50),cost_history[:minimum_index[0]+50],'b.')
ax.vlines(minimum_index[0], ymin = min(cost_history), ymax = max(cost_history),
           linestyles = '--', linewidth = 4, colors = 'green')
```

## 7.2 Mini Batch Gradient Descent
- Mini batch gradient descent approach uses a subset of samples randomly selected in batches (rather than all samples in gradient descent and a random sample in stochastic gradient descent). Thus, a faster optimization can be realized in practice since this approach only calculates the gradients for a group of observations.
- Steps (by each iteration)
    - 1. Shuffling data.
    - 2. Loop through shuffled data and select a batch of it.
    - 2. Compute the prediction given the batch of data.
    - 3. Update a subset of betas. 
    - 4. Each iteration ends until the program consumes all batches of data.

```python
def minibatch_gradient_descent(x,y,beta,learning_rate=0.01,iterations=10,batch_size=20):
    '''
    x: matrix of x with added bias units (beta 0), the shape must be 
    (number of data, number of predictors)
    y: vector of y, the shape must be (number of data, 1)
    beta: vector of betas, the shape must be (number of predictors, 1) 
    learning_rate: a float number 
    iterations: number of iterations
    batch_size: split data into batches, each batch contains batch_size of data points
    
    Returns the final beta vector and array of cost history.
    '''
    
    # number of data points
    m = len(y)
    
    # record the cost history 
    cost_history = np.zeros(iterations)
    
    # start iterations!
    for each_iter in range(iterations):
        cost = 0.0
        # shuffling data
        indices = np.random.permutation(m)
        x = x[indices]
        y = y[indices]
        # loop through all the data by the batch_size step
        for i in range(0,m,batch_size):
            # batch of data
            x_i = x[i:i+batch_size, :]
            y_i = y[i:i+batch_size, :]
            prediction = np.dot(x_i, beta)
            beta = beta -(1/m)*learning_rate*(x_i.T.dot((prediction - y_i)))
        # record the cost history for each iteration
        cost_history[each_iter] = cost_function(beta,x,y)
        
    return beta, cost_history
```

```python
# Start mini batch gradient descent!
beta, cost_history = minibatch_gradient_descent(X_b, y,
                                                beta=start_beta,
                                                learning_rate=0.1,
                                                iterations=1000)
```

```python
# Coefficients and MSE
print('Beta0:          {:0.3f},\nBeta1:          {:0.3f}'.format(beta[0][0],
                                                                 beta[1][0]))
print('Final cost/MSE:  {:0.3f}'.format(cost_history[-1]))
```
