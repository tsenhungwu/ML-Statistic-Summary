# Packages

```python
from scipy.stats import t
from scipy.stats import norm
import seaborn as sb
```

# T-Distribution

```python
# Create three random samples with different degree of freedom.
t_random_sample_1000_df_10 = t.rvs(loc = 0, scale = 1, df = 10, size = 1000)
t_random_sample_1000_df_30 = t.rvs(loc = 0, scale = 1, df = 30, size = 1000)
t_random_sample_1000_df_100 = t.rvs(loc = 0, scale = 1, df = 100, size = 1000)
```

```python
# Visualization 
sb.set(rc = {'figure.figsize': (11.7, 8.27)}, font_scale = 1.8)
ax = sb.distplot(t_random_sample_1000_df_10, kde = True, color = 'blue', hist_kws = {"linewidth": 25, 'alpha': 0.5})
ax.set(title = 'T-distribution with mean = 0, var = 1.25, d.f. = 10, and size = 1000', ylabel = 'Frequency')
```

```python
# Draw a random sample follows the normal distribution with the same parameters
n_random_sample_1000 = norm.rvs(loc = 0, scale = 1, size = 1000, random_state = 0)
```

```python
sb.set(rc = {'figure.figsize': (11.7, 8.27)}, font_scale = 1.6)
ax = sb.distplot(n_random_sample_1000, kde = True, hist = False, color = 'green', hist_kws = {"linewidth": 25, 'alpha': 0.2}, label = 'normal')
ax = sb.distplot(t_random_sample_1000_df_10, kde = True, hist = False, color = 'red', hist_kws = {"linewidth": 25, 'alpha': 0.2}, label = 't (10 d.f.)')
ax = sb.distplot(t_random_sample_1000_df_30, kde = True, hist = False, color = 'black', hist_kws = {"linewidth": 25, 'alpha': 0.2}, label = 't (30 d.f.)')
ax = sb.distplot(t_random_sample_1000_df_100, kde = True, hist = False, color = 'blue', hist_kws = {"linewidth": 25, 'alpha': 0.2}, label = 't (100 d.f.)')
ax.set(title = 'Normal VS T Distribution', ylabel = 'Frequency')
ax.legend()
```

# Statistics of T-Distribution

$$Var(X) = \frac {df} {df-2}$$

```python
t.var(df = 30, loc = 0, scale = 1)
```

# PDF & CDF of T-Distribution 
**(Compared with Normal Distribution)**

- **Normal Distribution**

```python
# The most three used alpha levels under normal distribution
# alpha = 0.01
print(round(norm.ppf(q = 0.995, loc = 0, scale = 1), 4))
print('alpha = 0.01:', round(norm.cdf(x = 2.5758, loc = 0, scale = 1), 4)) 
print()

# alpha = 0.05
print(round(norm.ppf(q = 0.975, loc = 0, scale = 1), 4))
print('alpha = 0.05:', round(norm.cdf(x = 1.96, loc = 0, scale = 1), 4)) 
print()

# alpha = 0.1
print(round(norm.ppf(q = 0.95, loc = 0, scale = 1), 4))
print('alpha = 0.1:', round(norm.cdf(x = 1.6449, loc = 0, scale = 1), 4))
```

```python
# The normal distribution has a bell symmetry shape.
# Thus, the probability (pdf) is the same given different directions but the same x variable.
print(round(norm.pdf(x = -1.96, loc = 0, scale = 1), 4)) 
print(round(norm.pdf(x = 1.96, loc = 0, scale = 1), 4)) 
```

- **T-Distribution**

```python
# All three levels have larger x variables compared with x variables in the normal distribution,
# since the t distribution has a fatter area in the tails and less area towards the middle part.
print(round(t.ppf(q = 0.995, df = 30, loc = 0, scale = 1), 4))
print('alpha = 0.01:', round(t.cdf(x = 2.75, df = 30, loc = 0, scale = 1), 4)) 
print()

print(round(t.ppf(q = 0.975, df = 30, loc = 0, scale = 1), 4))
print('alpha = 0.05:', round(t.cdf(x = 2.0423, df = 30, loc = 0, scale = 1), 4)) 
print()

print(round(t.ppf(q = 0.95, df = 30, loc = 0, scale = 1), 4))
print('alpha = 0.1:', round(t.cdf(x = 1.6973, df = 30, loc = 0, scale = 1), 4)) 
```

```python
# The t distribution has a bell symmetry shape, too.
# Thus, the probability (pdf) is the same given different directions but the same x variable.
print(round(t.pdf(x = -2.0423, df = 30, loc = 0, scale = 1), 4)) 
print(round(t.pdf(x = 2.0423, df = 30, loc = 0, scale = 1), 4)) 
```
