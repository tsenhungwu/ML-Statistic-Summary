# Packages

```python
import numpy as np
import seaborn as sb
import random
from scipy.stats import norm
```

# Normal Distribution

```python
# Specify mean and standard deviation
mu, sigma = 0, 0.1 
```

```python
# Draw a random sample with size = 100
s_100 = np.random.normal(mu, sigma, 100)
```

```python
# Verify the mean
print(abs(mu - np.mean(s_100)) < 0.01)

# Verify the std (ddof: degree of freedom for estimating the sample std)
print(abs(sigma - np.std(s_100, ddof=1)) < 0.01)
```

```python
# Using size = 100, the shape is not so close to the normal curve
sb.set(rc = {'figure.figsize': (11.7, 8.27)}, font_scale = 1.8)
ax = sb.distplot(s_100, kde = True, color = 'blue', hist_kws = {"linewidth": 25, 'alpha': 0.5})
ax.set(title = 'Normal distribution with size = 100', ylabel = 'Frequency')
```

```python
# Draw a random sample with size = 10000
s_10000 = np.random.normal(mu, sigma, 10000)
```

```python
# With size = 10000, the distribution is very close to a normal distribution
sb.set(rc = {'figure.figsize': (11.7, 8.27)}, font_scale = 1.8)
ax = sb.distplot(s_10000, kde = True, color = 'blue', hist_kws = {"linewidth": 25, 'alpha': 0.5})
ax.set(title = 'Normal distribution with size = 10000', ylabel = 'Frequency')
```

```python
# cdf
sb.set(rc = {'figure.figsize':(11.7, 8.27)}, font_scale = 1.8)
ax = sb.distplot(s_10000, kde = True, color = 'blue',\
                 hist_kws = dict(cumulative = True, alpha = 0.3),\
                 kde_kws = dict(cumulative = True))
ax.set(title = 'Normal cumulative distribution', ylabel = 'Frequency')
```

```python
# Random sampling from the population with three different sizes
sample_10 = random.sample(list(s_10000), 10)
sample_100 = random.sample(list(s_10000), 100)
sample_1000 = random.sample(list(s_10000), 1000)
```

```python
# As long as we increase the sample size larger, the closer to the original population shape 
sb.set(rc = {'figure.figsize': (11.7, 8.27)}, font_scale = 1.8)
ax = sb.distplot(s_10000, kde = True, color = 'blue', hist_kws = {"linewidth": 25, 'alpha': 0.2}, label = 'population')
ax = sb.distplot(sample_10, kde = True, color = 'red', hist_kws = {"linewidth": 25, 'alpha': 0.2}, label = '10 sample')
ax = sb.distplot(sample_100, kde = True, color = 'yellow', hist_kws = {"linewidth": 25, 'alpha': 0.2}, label = '100 sample')
ax = sb.distplot(sample_1000, kde = True, color = 'green', hist_kws = {"linewidth": 25, 'alpha': 0.2}, label = '1000 sample')
ax.set(title = 'Compare diffent sample sizes draw from the population', ylabel = 'Frequency')
ax.legend()
```

# Statistics of normal distribution

```python
# mean, variance, and std
print(norm.mean(loc=0, scale=2))
print(norm.var(loc=0, scale=2))
print(norm.std(loc=0, scale=2))
```

#  PDF & CDF of normal distribution

```python
# Probability density function
print(round(norm.pdf(1.645, loc=0, scale=1), 4))
print(round(norm.pdf(1.96, loc=0, scale=1), 4))
print(round(norm.pdf(3, loc=0, scale=1), 4))
```

```python
# Cumulative density function
print(round(norm.cdf(1.645, loc=0, scale=1), 4))
print(round(norm.cdf(1.96, loc=0, scale=1), 4))
print(round(norm.cdf(3, loc=0, scale=1), 4))
```
