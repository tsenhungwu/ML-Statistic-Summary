# Binomial distribution

```python
# Self-defined bernoulli function
def self_binomial(success_p, total_event, success_event):
    '''
    success_p: success probability of an event, float number
    total_event: indicate how many total events, integer number
    success_event: indicate how many successful events, integer number
    '''
    failure_event = total_event - success_event
    combination = comb(total_event, success_event)
    
    return combination*((success_p**(success_event))*((1-success_p)**(total_event-success_event)))
```

```python
# For instance, the success probability is 0.4
# Given 5 total events included 3 successful events, we can calculate the corresponding probability
print('The probability is', round(self_binomial(success_p = 0.4, total_event = 5, success_event = 3),4),\
      ', given 5 total events included 3 successful events and success probability 0.4.')
```
