# Packages

```python
from scipy.stats import binom
from scipy.special import comb
import seaborn as sb
```

# Binomial distribution

```python
# Self-defined bernoulli probability mass function
def self_binomial(success_p, total_event, success_event):
    '''
    success_p: success probability of an event, float number (from 0 to 1)
    total_event: indicate how many total events, integer number
    success_event: indicate how many successful events, integer number (must less than total events)
    '''
    failure_event = total_event - success_event
    combination = comb(total_event, success_event)
    
    return combination*((success_p**(success_event))*((1-success_p)**(total_event-success_event)))
```

```python
# For instance, the success probability is 0.4.
# Given number of total events, number of successful events, and success probability, 
# we can calculate the corresponding probability.
print('The probability is', round(self_binomial(success_p = 0.4, total_event = 5, success_event = 3),4),\
      ', given 5 total events included 3 successful events and success probability 0.4.')
```

```python
# Or, use the binom module 
# n = total number of event
# k = number of successful event
# p = successful probability of an event 
binom.pmf(k = 3, n = 5, p = 0.4)
```

# Statistics of Binomial distribution

```python
print('Median is', binom.median(n = 5, p = 0.4), ', given the success probability 0.4 and total number of event 5.')
```

```python
# Definition: mean = n * p
print('Mean is', binom.mean(n = 5, p = 0.4), ', given the success probability 0.4 and total number of event 5.')
```

```python
# Definition: variance = np * (1-p)
print('Variance is', binom.var(n = 5, p = 0.4), ', given the success probability 0.4 and total number of event 5.')
```

```python
# Definition: standard deviation = variance ** (1/2)
print('Standard deviation is', binom.std(n = 5, p = 0.4), ', given the success probability 0.4 and total number of event 5.')
print(binom.std(n = 5, p = 0.4) == binom.var(n = 5, p = 0.4) ** (1/2))
```

# Visualization of Binomial distribution

```python
# Notice that for a small number of experiment, the mean is 15 (we expect the mean should be 20 * 0.8)
# Compared the result with the below experiment using large number
sb.set(rc = {'figure.figsize': (11.7, 8.27)}, font_scale = 1.8)
data = binom.rvs(n = 20, p = 0.8, loc = 0, size = 10)
ax = sb.distplot(data, kde = True, color = 'blue', hist_kws = {"linewidth": 25, 'alpha': 0.5})
ax.set(title = 'Binomial distribution', ylabel = 'Frequency')
```

```python
sb.set(rc = {'figure.figsize':(11.7, 8.27)}, font_scale = 1.8)
ax = sb.distplot(data, kde = True, color = 'blue',\
                 hist_kws = dict(cumulative = True, alpha = 0.3),\
                 kde_kws = dict(cumulative = True))
ax.set(title = 'Binomial cumulative distribution', ylabel = 'Frequency')
```

```python
# The mean here is 16. (20 * 0.8)
sb.set(rc = {'figure.figsize':(11.7, 8.27)}, font_scale = 1.8)
data = binom.rvs(n = 20,p = 0.8, loc = 0, size = 1000)
ax = sb.distplot(data, kde = True, color = 'blue', hist_kws = {"linewidth": 25, 'alpha': 0.5})
ax.set(title = 'Binomial distribution', ylabel = 'Frequency')
```

```python
# After 16, the cumulative probability is over 50%!
sb.set(rc = {'figure.figsize':(11.7, 8.27)}, font_scale = 1.8)
ax = sb.distplot(data, kde = True, color = 'blue',\
                 hist_kws = dict(cumulative = True, alpha = 0.3),\
                 kde_kws = dict(cumulative = True))
ax.set(title = 'Binomial cumulative distribution', ylabel = 'Frequency')
```
