# Packages

```python
from scipy.stats import bernoulli
import seaborn as sb
```

# Bernoulli distribution

```python
# Self-defined bernoulli function
def self_bernoulli(success_p, success = True):
    '''
    success_p: success probability of an event, float number
    success: indicate whether an event is successful or fail, boolean values
    '''
    return (success_p**(success))*((1-success_p)**(1-success))
```

```python
# For instance, if the success probability is 0.4 and therefore we get the success probability 0.4.
# Otherwise, the probability is 0.6 for failure.
print('Success probability is',self_bernoulli(0.4, success = True))
print('Failure probability is',self_bernoulli(0.4, success = False))
```

```python
# Or can use the bernoulli module
print(bernoulli.pmf(p = 0.22, k = 0))
print(bernoulli.pmf(p = 0.22, k = 1))

print(bernoulli.pmf(p = 0.4, k = 0))
print(bernoulli.pmf(p = 0.4, k = 1))
```

# Statistics of Bernoulli distribution

```python
print('Median is', bernoulli.median(0.6), 'given the success probability 0.6.')
```

```python
# Definition: mean (or expectation) = p
print('Mean is', bernoulli.mean(0.6), 'given the success probability 0.6.')
```

```python
# Definition: variance = p * (1-p)
print('Variance is', bernoulli.var(0.6), 'given the success probability 0.6.')
```

```python
# Definition: standard deviation = variance ** (1/2)
print('Standard deviation is', bernoulli.std(0.6), 'given the success probability 0.6.')
print(bernoulli.std(0.6) ==  bernoulli.var(0.6) ** (1/2))
```

# Visualization of Bernoulli distribution

```python
# Visualization of random 50 data draw from a bernoulli distribution
data = bernoulli.rvs(size=50, p=0.6)
ax = sb.distplot(data, kde=True, color='crimson', hist_kws={"linewidth": 25,'alpha':1})
ax.set(title='Bernouli Distribution', ylabel='Frequency')
```
