# Linear Congruential Method

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc = {'figure.figsize':(11.7, 8.27)}, font_scale = 1.8)
```

```python
def generate_lcg(num_iterations, verbose = None):
    """
    LCG - generates as many random numbers as requested by user, using a Linear Congruential Generator
    LCG uses the formula: X_(i+1) = (aX_i + c) mod m
    :param num_iterations: int - the number of random numbers requested
    :return: void
    """
    # Initialize variables
    x_value = 123456789.0    # Initial seed, or X_0 = 123456789
    a = 101427               # Our "a" base value
    c = 321                  # Our "c" base value
    m = (2 ** 16)            # Our "m" base value

    # Counter for how many iterations we've run
    counter = 0

    # Create an empty list to collect random numbers (from 0 to 1)
    random_list = []

    # Perfom number of iterations requested by user
    while counter < num_iterations:
        # Store value of each iteration
        x_value = (a * x_value + c) % m

        # Obtain each number in U[0,1) by diving X_i by m
        random_number = (x_value/m)

        # Collect results
        random_list.append(random_number)

        counter = counter + 1
        
    if verbose == True:
        print("Successfully generated " + str(num_iterations) + " random numbers.")
        
    return random_list
```

```python
# Perform 3 iterations
generate_lcg(3)
```

# Problem: poor choice of the constants can lead to very poor sequence

```python
def poor_generate_lcg(num_iterations, verbose = None):
    """
    LCG - generates as many random numbers as requested by user, using a Linear Congruential Generator
    LCG uses the formula: X_(i+1) = (aX_i + c) mod m
    :param num_iterations: int - the number of random numbers requested
    :return: void
    """
    # Initialize variables
    x_value = 7   
    a = 5          
    c = 3          
    m = 16         

    # Counter for how many iterations we've run
    counter = 0

    # Create an empty list to collect random numbers (from 0 to 1)
    random_list = []

    # Perfom number of iterations requested by user
    while counter < num_iterations:
        # Store value of each iteration
        x_value = (a * x_value + c) % m

        # Obtain each number in U[0,1) by diving X_i by m
        random_number = (x_value/m)

        # Collect results
        random_list.append(random_number)

        counter = counter + 1
        
    if verbose == True:
        print("Successfully generated " + str(num_iterations) + " random numbers.")
        
    return random_list
```

```python
# Use a poor generator
x = [i for i in range(32)]
plt.figure(figsize=(20,10))
plt.plot(x, poor_generate_lcg(32))
plt.title('16 iterations as one cycle', fontsize = 30)
plt.axvline(x = 16, c = 'red')
```

```python
# Certain pattern found!
plt.figure(figsize=(20,10))
sns.distplot(poor_generate_lcg(1000))
```

```python
# Use a favorable generator
x = [i for i in range(32)]
plt.figure(figsize=(20,10))
plt.plot(x, generate_lcg(32))
```
