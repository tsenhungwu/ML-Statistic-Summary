```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc = {'figure.figsize':(11.7, 8.27)}, font_scale = 1.8)
```

# RANDU Generator

```python
def generate_randu(num_iterations, verbose = None):
    """
    The difference in randu is that it doesn't contain c parameter.
    """
    # Initialize variables
    x_value = 1            # Initial seed, or X_0 = 1
    a = 7 ** 5             # Our "a" base value
    m = (2 ** 31) - 1      # Our "m" base value

    # Counter for how many iterations we've run
    counter = 0

    # Create an empty list to collect random numbers (from 0 to 1)
    random_list = []

    # Perfom number of iterations requested by user
    while counter < num_iterations:
        # Store value of each iteration
        x_value = (a * x_value ) % m

        # Obtain each number in U[0,1) by diving X_i by m
        random_number = (x_value / m)

        # Collect results
        random_list.append(random_number)

        counter = counter + 1
        
    if verbose == True:
        print("Successfully generated " + str(num_iterations) + " random numbers.")
        
    return random_list
```

```python
# Run 5 iterations
generate_randu(5)
```

```python
# Somehow not uniformly distributed
plt.figure(figsize=(20,10))
sns.distplot(generate_randu(1000))
```

```python
# Increase the size to 10000, more uniform distribution
# The choice of a & m is critical!
plt.figure(figsize=(20,10))
sns.distplot(generate_randu(10000))
```

# 1960's IBM RANDU Generator
